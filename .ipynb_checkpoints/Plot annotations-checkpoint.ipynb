```python
import os
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
import numpy as np
import re
plt.style.use('seaborn-notebook')
SMALL_SIZE = 16
MEDIUM_SIZE = 18
BIGGER_SIZE = 20

plt.rc('font', size=SMALL_SIZE)          # controls default text sizes
plt.rc('axes', titlesize=SMALL_SIZE)     # fontsize of the axes title
plt.rc('axes', labelsize=MEDIUM_SIZE)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('ytick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('legend', fontsize=SMALL_SIZE)    # legend fontsize
plt.rc('figure', titlesize=BIGGER_SIZE)  # fontsize of the figure title
```

```python
os.chdir('annotations/')
annotations=os.listdir()
```

```python
def get_spindles(x):
    if re.compile('spindle',re.IGNORECASE).search(x):
        return True
    else:
        return False
```

```python
fig, ax = plt.subplots(nrows=2,figsize=(16,16))
idx1=1;idx2=1;
subplot1_yticks,subplot1_yticklabels,subplot2_yticks,subplot2_yticklabels=[],[],[],[]
sub_20, sub_30 = [],[]
for cnt,annotation in enumerate(annotations):
    if re.compile('nap',re.IGNORECASE).search(annotation):
        working_df = pd.read_csv(annotation)
        spindle_df = working_df[working_df['Annotation'].apply(get_spindles)]
        ax[0].scatter(spindle_df['Onset'],(idx1)*np.ones(spindle_df['Onset'].values.size))
        subplot1_yticks.append(idx1)
        subplot1_yticklabels.append(annotation.split('_')[0]+annotation.split('_')[2])
        sub_20.append(spindle_df['Onset'].values)
        idx1 += 1
    else:
        working_df = pd.read_csv(annotation)
        spindle_df = working_df[working_df['Annotation'].apply(get_spindles)]
        ax[1].scatter(spindle_df['Onset'],(idx2)*np.ones(spindle_df['Onset'].values.size))
        subplot2_yticks.append(idx2)
        subplot2_yticklabels.append(annotation.split('_')[0]+annotation.split('_')[1])
        sub_30.append(spindle_df['Onset'].values)
        idx2 += 1
_=ax[0].set(yticks=subplot1_yticks,yticklabels=subplot1_yticklabels,
           xlim=(-0.5,2000),ylim=(0.5,len(subplot1_yticks)+0.5),title='Sleep for 20 minutes',
           xlabel='Time (sec)',ylabel='Subject and day')
_=ax[1].set(yticks=subplot2_yticks,yticklabels=subplot2_yticklabels,
           xlim=(-0.5,3000),ylim=(0.5,len(subplot2_yticks)+0.5),title='Sleep for 30 minutes',
           xlabel='Time (sec)',ylabel='Subject and day')
```

```python
import pickle

with open('old_subjects.p', 'wb') as fp:
    pickle.dump(sub_30, fp)
with open('new_subjects.p','wb') as fp:
    pickle.dump(sub_20,fp)
```
